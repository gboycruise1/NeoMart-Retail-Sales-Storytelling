```python
#Turning into Date column
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
```

```python
df["Month"] = df["Order_Date"].dt.month_name()
df["Month_No"] = df["Order_Date"].dt.month
df["Quarter"] = df["Order_Date"].dt.quarter
df["Year"] = df["Order_Date"].dt.year
```

### Observations

- The dataset contains 2000 rows and 18 columns.
- There are no missing values 0 
- There are no duplicate records 
- The Order_Date column was converted to datetime format.
- New features (Month, Month_No, Quarter, and Year) were created to support time-based analysis.

## KPI


```python
total_sales = df["Sales"].sum()
print(f"Total Sales: ₦{total_sales:,.2f}")
```

```python
total_profit = df["Profit"].sum()
print(f"Total Profit: ₦{total_profit:,.2f}")
```

```python
avg_satisfaction = df["Customer_Satisfaction"].mean()
print(f"Average Customer Satisfaction: {avg_satisfaction:.2f}")
```

```python
avg_delivery = df["Delivery_Days"].mean()
print(f"Average Delivery Time: {avg_delivery:.2f} days")
```

```python
region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)

print(region_sales)

highest_region = region_sales.idxmax()

print(f"\nHighest Performing Region: {highest_region}")
```

```python
lowest_region = region_sales.idxmin()

print(f"Lowest Performing Region: {lowest_region}")
```

```python
category_profit = df.groupby("Product_Category")["Profit"].sum().sort_values(ascending=False)

print(category_profit)

best_category = category_profit.idxmax()

print(f"\nBest Product Category: {best_category}")
```

```python
worst_category = category_profit.idxmin()

print(f"Worst Product Category: {worst_category}")
```

```python
kpi_summary = pd.DataFrame({
    "KPI": [
        "Total Sales",
        "Total Profit",
        "Average Customer Satisfaction",
        "Average Delivery Time",
        "Highest Performing Region",
        "Lowest Performing Region",
        "Best Product Category",
        "Worst Product Category"
    ],
    "Value": [
        f"₦{total_sales:,.2f}",
        f"₦{total_profit:,.2f}",
        f"{avg_satisfaction:.2f}",
        f"{avg_delivery:.2f} days",
        highest_region,
        lowest_region,
        best_category,
        worst_category
    ]
})

kpi_summary
```

## KPI Insights

- Total sales indicate strong revenue generation during the year.
- Profit is positive but should be compared with sales trends to evaluate profitability.
- The highest-performing region contributed the largest share of sales.
- The lowest-performing region may require additional marketing or operational improvements.
- The best-performing product category generated the highest total profit.
- Customer satisfaction and delivery performance provide insight into overall customer experience.

```python
monthly_sales = (
    df.groupby(["Month_No", "Month"])["Sales"]
    .sum()
    .reset_index()
    .sort_values("Month_No")
)

plt.figure(figsize=(10,5))
plt.plot(monthly_sales["Month"], monthly_sales["Sales"], marker="o", linewidth=2)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)

plt.show()
```

Sales fluctuated throughout the year. The chart helps identify peak and low sales months for better planning.
 We had a peak sale in the month of December

```python
monthly_profit = (
    df.groupby(["Month_No", "Month"])["Profit"]
    .sum()
    .reset_index()
    .sort_values("Month_No")
)

plt.figure(figsize=(10,5))
plt.plot(monthly_profit["Month"], monthly_profit["Profit"], marker="o", color="green", linewidth=2)

plt.title("Monthly Profit Trend")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.xticks(rotation=45)

plt.show()
```

Profit is not increasing at the same rate as sales.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create figure
fig = plt.figure(figsize=(18, 12))

# Dashboard title
fig.suptitle(
    "NeoMart Retail Sales Executive Dashboard",
    fontsize=22,
    fontweight="bold"
)

# ------------------------
# KPI Cards
# ------------------------

plt.figtext(
    0.10, 0.90,
    f"Total Sales\n₦{total_sales:,.2f}",
    fontsize=14,
    bbox=dict(facecolor="#D6EAF8", edgecolor="black", boxstyle="round,pad=0.8")
)

plt.figtext(
    0.33, 0.90,
    f"Total Profit\n₦{total_profit:,.2f}",
    fontsize=14,
    bbox=dict(facecolor="#D5F5E3", edgecolor="black", boxstyle="round,pad=0.8")
)

plt.figtext(
    0.56, 0.90,
    f"Avg Satisfaction\n{avg_satisfaction:.2f}/5",
    fontsize=14,
    bbox=dict(facecolor="#FCF3CF", edgecolor="black", boxstyle="round,pad=0.8")
)

plt.figtext(
    0.79, 0.90,
    f"Avg Delivery\n{avg_delivery:.2f} Days",
    fontsize=14,
    bbox=dict(facecolor="#FADBD8", edgecolor="black", boxstyle="round,pad=0.8")
)

# ------------------------
# Charts
# ------------------------

ax1 = plt.subplot(2,2,1)

ax1.plot(
    monthly_sales["Month"],
    monthly_sales["Sales"],
    marker="o"
)
ax1.set_title("Monthly Sales Trend")
ax1.tick_params(axis='x', rotation=45)

# ------------------------

ax2 = plt.subplot(2,2,2)

ax2.plot(
    monthly_profit["Month"],
    monthly_profit["Profit"],
    marker="o",
    color="green"
)
ax2.set_title("Monthly Profit Trend")
ax2.tick_params(axis='x', rotation=45)

# ------------------------

ax3 = plt.subplot(2,2,3)

region_sales = df.groupby("Region")["Sales"].sum()

ax3.bar(region_sales.index, region_sales.values)

ax3.set_title("Regional Sales")

# ------------------------

ax4 = plt.subplot(2,2,4)

sns.boxplot(
    data=df,
    x="Region",
    y="Customer_Satisfaction",
    ax=ax4
)

ax4.set_title("Customer Satisfaction by Region")

plt.tight_layout(rect=[0,0,1,0.88])

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# KPI Calculations
# -----------------------------
total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()
avg_satisfaction = df["Customer_Satisfaction"].mean()
avg_delivery = df["Delivery_Days"].mean()

region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
highest_region = region_sales.idxmax()
lowest_region = region_sales.idxmin()

category_profit = df.groupby("Product_Category")["Profit"].sum().sort_values(ascending=False)
best_category = category_profit.idxmax()
worst_category = category_profit.idxmin()

monthly_sales = (
    df.groupby(["Month_No", "Month"])["Sales"]
    .sum()
    .reset_index()
    .sort_values("Month_No")
)

monthly_profit = (
    df.groupby(["Month_No", "Month"])["Profit"]
    .sum()
    .reset_index()
    .sort_values("Month_No")
)

# -----------------------------
# Dashboard
# -----------------------------
fig = plt.figure(figsize=(24,28))

fig.suptitle(
    "NeoMart Retail Sales Executive Dashboard",
    fontsize=24,
    fontweight="bold"
)

# -----------------------------
# KPI Cards
# -----------------------------

kpis = [
    ("Total Sales", f"₦{total_sales:,.0f}", 0.03),
    ("Total Profit", f"₦{total_profit:,.0f}", 0.15),
    ("Avg Satisfaction", f"{avg_satisfaction:.2f}/5", 0.27),
    ("Avg Delivery", f"{avg_delivery:.2f} Days", 0.39),
    ("Best Region", highest_region, 0.51),
    ("Worst Region", lowest_region, 0.63),
    ("Best Category", best_category, 0.75),
    ("Worst Category", worst_category, 0.87),
]

for title, value, xpos in kpis:
    plt.figtext(
        xpos,
        0.94,
        f"{title}\n{value}",
        ha="center",
        fontsize=12,
        bbox=dict(
            facecolor="lightblue",
            edgecolor="black",
            boxstyle="round,pad=0.6"
        )
    )

# -----------------------------
# 8 Charts
# -----------------------------

gs = fig.add_gridspec(
    4,
    2,
    left=0.05,
    right=0.95,
    bottom=0.05,
    top=0.88,
    hspace=0.45,
    wspace=0.25
)

# -----------------------------
# 1 Monthly Sales
# -----------------------------
ax1 = fig.add_subplot(gs[0,0])

ax1.plot(
    monthly_sales["Month"],
    monthly_sales["Sales"],
    marker="o",
    linewidth=3
)

ax1.set_title("Monthly Sales Trend")
ax1.tick_params(axis="x", rotation=45)

# -----------------------------
# 2 Monthly Profit
# -----------------------------
ax2 = fig.add_subplot(gs[0,1])

ax2.plot(
    monthly_profit["Month"],
    monthly_profit["Profit"],
    marker="o",
    linewidth=3,
    color="green"
)

ax2.set_title("Monthly Profit Trend")
ax2.tick_params(axis="x", rotation=45)

# -----------------------------
# 3 Sales by Region
# -----------------------------
ax3 = fig.add_subplot(gs[1,0])

sns.barplot(
    data=df,
    x="Region",
    y="Sales",
    estimator=sum,
    errorbar=None,
    ax=ax3
)

ax3.set_title("Sales by Region")

# -----------------------------
# 4 Profit by Category
# -----------------------------
ax4 = fig.add_subplot(gs[1,1])

sns.barplot(
    data=df,
    x="Product_Category",
    y="Profit",
    estimator=sum,
    errorbar=None,
    ax=ax4
)

ax4.set_title("Profit by Product Category")
ax4.tick_params(axis="x", rotation=20)

# -----------------------------
# 5 Discount vs Profit
# -----------------------------
ax5 = fig.add_subplot(gs[2,0])

sns.scatterplot(
    data=df,
    x="Discount",
    y="Profit",
    ax=ax5
)

ax5.set_title("Discount vs Profit")

# -----------------------------
# 6 Correlation Heatmap
# -----------------------------
ax6 = fig.add_subplot(gs[2,1])

corr = df.select_dtypes(include="number").corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    ax=ax6
)

ax6.set_title("Correlation Heatmap")

# -----------------------------
# 7 Customer Satisfaction
# -----------------------------
ax7 = fig.add_subplot(gs[3,0])

sns.boxplot(
    data=df,
    x="Region",
    y="Customer_Satisfaction",
    ax=ax7
)

ax7.set_title("Customer Satisfaction by Region")

# -----------------------------
# 8 Delivery Performance
# -----------------------------
ax8 = fig.add_subplot(gs[3,1])

sns.boxplot(
    data=df,
    x="Region",
    y="Delivery_Days",
    ax=ax8
)

ax8.set_title("Delivery Performance by Region")

# -----------------------------
# Save Dashboard
# -----------------------------
plt.savefig(
    "../dashboard.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
